```python
ss=x.copy()
ss["date"]=pd.to_datetime(ss["time_stamp"])
ss.head
```

```python
ss.describe().T
```

```python
def rolling_count(ss_group, frequency):
    return ss_group.rolling(frequency, on='date')['user_id'].count()
ss['visits_7_days'] = ss.groupby('user_id', as_index=False, group_keys=False).apply(rolling_count, '7D')
```

```python
ss.head()
```

```python
ss[ss["visits_7_days"] >= 3.0]
```

```python
user_adopted = ss.groupby('user_id')['visits_7_days'].max().reset_index()
user_adopted
```

```python
user_adopted['adopted_user'] = user_adopted['visits_7_days'].apply(lambda x: 1 if x>=3 else 0)
user_adopted.head()
```
